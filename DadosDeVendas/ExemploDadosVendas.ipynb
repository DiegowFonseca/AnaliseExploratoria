## Sample Sales Data

- Conjunto de dados do site Keggle: https://www.kaggle.com/datasets/kyanyoga/sample-sales-data
- O conjunto de dados se trata de vendas de brinquedos para algumas lojas ao redor do mundo.

### Entendendo a base de dados

```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importando a base de dados
base = pd.read_csv('sales_data_sample.csv', encoding="ISO-8859-1")
base
```

```python
base.info()
```

```python
# Verificando os valores nulos
base.isnull().sum()
```

```python
# Olhando quais paises possui na coluna TERRITORY valores nulos
base.groupby('COUNTRY').count()
```

### Limpando e pré-processando os dados

- Nessa etapa vou apagar as colunas 'ADDRESSLINE2', 'STATE' e 'TERRITORY' por ter muitos valores ausentes. 
- Vou apagar as colunas: 'PHONE', 'STATUS', 'ADDRESSLINE1', 'POSTALCODE', 'CONTACTLASTNAME' e 'CONTACTFIRSTNAME' que não serão necessárias para a análise que irei fazer.

```python
# Elimiando colunas desnecessárias
base = base.drop(['ADDRESSLINE1', 'ADDRESSLINE2','STATE','TERRITORY','PHONE','STATUS','POSTALCODE','CONTACTLASTNAME','CONTACTFIRSTNAME'], axis=1)
```

```python
# Verificando se possui alguma coluna com valores ausentes
base.isnull().sum()
```

```python
# Olhando se tem valores duplicados
base.duplicated().sum()
```

```python
# Transformar a coluna ORDERDATE em datetime
base['ORDERDATE'] = pd.to_datetime(base['ORDERDATE'], format='%m/%d/%Y %H:%M')
```

```python
# Criando a coluna de ANO
base['YEAR'] = base['ORDERDATE'].dt.year

# Criando a coluna de MES
base['MONTH'] = base['ORDERDATE'].dt.month
```

### Análise Exploratória de Dados

```python
# Visualizando dados estatísticos
base.describe()
```

- Distribuição de vendas de acordo com a linha do produto

```python
sns.set_theme(palette='pastel')

sns.boxplot(x=base['SALES'], y= base['PRODUCTLINE'], palette=['m', 'g'])
plt.show()
```

- Quantidade encomendada pela linha de produto

```python
base.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().plot(kind='bar')

plt.title('Quantidade Encomendada por Linha de Produtos')
plt.ylabel('Quantidade Total Encomendada')
plt.show()
```

- Total de vendas de acordo com a linha do produto

```python
base.groupby('PRODUCTLINE')['SALES'].sum().plot(kind='bar')

plt.title('Total de Vendas por Linha de Produtos')
plt.ylabel('Total de Vendas')

plt.show()
```

- Total de vendas de acordo com cada ano

```python
base.groupby('YEAR')['SALES'].sum().plot(kind='bar')

plt.title('Total de Vendas por Ano')
plt.ylabel('Total de Vendas')
plt.show()
```

```python
base.loc[base['YEAR'] == 2005, 'MONTH'].max()  
```

Como em 2005 só teve vendas até o mês 5, eu criei um dataframe contendo somente as vendas até o mês de maio de cada ano.

```python
ano2003 = base.loc[((base['YEAR'] == 2003) & (base['MONTH']<=5)), :]
ano2004 = base.loc[((base['YEAR']==2004) & (base['MONTH']<=5)), :]
ano2005 = base.loc[((base['YEAR']==2005) & (base['MONTH']<=5)), :]

anos = pd.concat([ano2003, ano2004, ano2005])
anos.head()
```

Foi possível perceber que o ano de 2005 até o mês de maio as vendas foram superiores aos outros anos.

```python
anos.groupby('YEAR')['SALES'].sum().plot(kind='bar')

plt.title('Total de Vendas por Ano')
plt.ylabel('Total de Vendas')

plt.show()
```

- Vou avaliar agora todo o ano de 2003 e 2004 para ver se possui um período do ano onde acontece mais vendas.

```python
# Olhando a evolução das vendas ao longo dos meses no ano de 2003
ano1 = base.loc[base['YEAR']==2003,:]
ano1.groupby('MONTH')['SALES'].sum().plot(kind='bar')
plt.show()
```

```python
# Olhando a evolução de vendas ao longo dos meses no ano de 2004
ano2 = base.loc[base['YEAR']==2004,:]
ano2.groupby('MONTH')['SALES'].sum().plot(kind='bar')
plt.show()
```

Esse aumento nas vendas no mês 11 pode ser explicado porque dia 20 de novembro a ONU estabeleceu como o Dia Mundial das Crianças. Esse período as crianças tendem a ganhar muitos presentes, inclusive brinquedos.

- Quem são os clientes mais frequentes e valiosos?

```python
# De acordo com a quantidade encomendada
b1 = base.groupby(['CUSTOMERNAME'])[['QUANTITYORDERED']].sum().reset_index().sort_values(by='QUANTITYORDERED', ascending=False)[:5]
b1
```

```python
# De acordo com a quantidade de vendas
b2 = base.groupby(['CUSTOMERNAME'])[['SALES']].sum().reset_index().sort_values(by='SALES', ascending=False)[:5]
b2
```

- Total de vendas pelo tamanho da empresa

```python
b_dealsize = base.groupby('DEALSIZE')[['SALES']].sum().sort_values(by='SALES')

b_dealsize.plot(kind='barh')
plt.xlabel('Total de Vendas')
plt.ylabel('Tamanho da Empresa')
plt.show()
```

- De quais países mais teve vendas e quantidades encomendadas?

```python
b_vendas = base.groupby('COUNTRY')[['SALES']].sum().sort_values(by='SALES')

b_vendas.plot(kind='barh')

plt.title('Total de Vendas por País')
plt.show()
```

```python
b_encomenda = base.groupby('COUNTRY')[['QUANTITYORDERED']].sum().sort_values(by='QUANTITYORDERED')

b_encomenda.plot(kind='barh')

plt.title('Quantidade Encomendada por País')
plt.show()
```

- Quais foram as 5 cidades onde teve mais vendas?

```python
city = base.groupby(['CITY'])[['SALES']].sum().reset_index().sort_values(by='SALES', ascending=False)[:5]
city
```

### Conclusão

- Tem aumentado o número de vendas ao longo dos anos, e o período onde tem mais venda é no mês de novembro, que é considerado o mês Mundial das Crianças, onde muitos brinquedos são comprados para presentear as crianças.
- Você tem os Estados Unidos liderando nas vendas, mas a cidade que mais vendeu foi Madrid que pertence a Espanha e Nova York ficou em terceiro lugar.
- As maiores vendas foram feitas para empresas de médio porte.
